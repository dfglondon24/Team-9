```python
import pandas as pd 
import numpy as np
import re
```

```python
df = pd.read_csv('challenges_data.csv')
```

```python
df.head()
```

```python
df['Grade'] = df['Grades'].apply(lambda x: min(map(int, re.findall(r'\d+', str(x)))) if re.findall(r'\d+', str(x)) else None)
df
```

```python
df.dropna(how='all', inplace=True)
df = df.rename(columns={"1.Name of District": "Name of District", "2.Name of Community":"Name of Community", "4. Sex": "Sex", "5. Age(in years)": "Age", "6. Educational Level": "Educational Level"})
```

```python
df.drop(columns={'Grades'}, inplace=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

plt.figure(figsize=(10, 6))
sns.boxplot(x='Organisation helping child abuse', y='Grade', data=df)

# Add title and labels
plt.title('Grade Distribution across Organisations Helping Child Abuse')
plt.xlabel('Organisation Helping Child Abuse')
plt.ylabel('Grade')

# Display the plot
plt.show()
```

```python
afrikids_grades = df[df['Organisation helping child abuse'] == 'AfriKids']['Grade']
other_grades = df[df['Organisation helping child abuse'] != 'AfriKids']['Grade']

plt.hist(afrikids_grades, alpha=0.5, label='AfriKids', bins=10)
plt.hist(other_grades, alpha=0.5, label='Other Organizations', bins=10)
plt.legend(loc='upper right')
plt.title('Grade Distribution: Afrikids vs Other Organizations')
plt.show()
```

```python
df_afrikids = df[df['Organisation helping child abuse'] == 'AfriKids']

order = ['JHS', 'Upper primary', 'Lower primary']

sns.boxplot(x='Educational Level', y='Grade', hue='Organisation helping child abuse', data=df_afrikids, palette='Greens', order=order)
plt.title('Grade Distribution by Educational Level with AfriKids Intervention')
plt.legend([],[], frameon=False)

plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

custom_stopwords = set(STOPWORDS)  # Add default stopwords
custom_stopwords.update(['text', 'book', 'challenge', 'challenges', 'books', 'lacks', 'lack', 'inadequate'])

# Combine all text in the 'Challenges' column
challenges_text = " ".join(review for review in df['Challenges'].astype(str))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=custom_stopwords).generate(challenges_text)


# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No axes for a cleaner look
plt.title('Word Cloud for Challenges Faced by Students', fontsize=15)
plt.show()

```
